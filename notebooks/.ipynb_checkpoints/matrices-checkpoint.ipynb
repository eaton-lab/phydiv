```python
#adding probabilities to distances

#-1 and 1 are max probs (p = 1 [needs to be 0.9999999]), and 0 is 50/50 (p = 0.5)
p = 0.6
max_p = abs(p)/2 + 0.5
min_p = 1 - max_p
print(f"min = {min_p}, max = {max_p}")

#re-scaling distances to probabilities
dist_ex = [0.06, 0.5, 0.9, 1.2, 1.8, 2]
new_dist = []
for pr in dist_ex:
    new_p = min_p + (pr - min(dist_ex)) * (max_p - min_p) / (max(dist_ex) - min(dist_ex))
    new_dist.append(new_p)
#print(dist_ex, new_dist)

#using indexing
tip_dists = dm.loc[tip].copy()
probabilities = min_p + (tip_dists - min(tip_dists)) * (max_p - min_p) / (max(tip_dists) - min(tip_dists))
print(tip_dists)
print(probabilities)
```

```python
def simcom(sr, pa = 0.5, verbose = False):
    """
    Simulate a community from the species pool under different phylogenetic structure assumptions

    Parameters:
    ---
    sr: int; species richness
        - Must be int < len(species)
    pa: float (-1 to 1); phylogenetic assumption
        - -1 = related species are least likely to co-occur
        - 0 = no phylogenetic structure
        - 1 = related species are most likely to co-occur

    Return:
    ---
    matrix of species presence/absence
    
    """

    #Get species distance matrix
    dm = sp_tree.distance.get_tip_distance_matrix()

    #Start community with a random species
    comm = []
    start = np.random.choice(dm.index)
    comm.append(start)
    if verbose:
        print(f"The starting species is {comm}")

    #Making a list of unselected species
    unused = set(dm.index)
    unused.remove(start)

    #Creating probability weights (min and max are 0.5 for pa = 0.5)
    if abs(pa) == 1:
        raise Exception("abs(pa) = 1 may cause too few species in the community")
    max_p = abs(pa)/2 + 0.5
    min_p = 1 - max_p
    if verbose:
        print(f"min = {min_p}, max = {max_p}")
    
    #Select species from the distance matrix
    for sp in (sr - 1):
        #stop if no more species remain
        if not unused:
            raise Exception("Species richness exceeds the species pool. No more species are unrepresented.")
            break

        #get distances for current species
        current_sp = comm[-1]
        sp_dists = dm.loc[current_sp].copy()

        #setting already selected species distances to zero
        for name in dm.index:
            if name not in unused:
                sp_dists[name] = 0

        #apply weighted probabilities to distances
        if sp_dists.sum() > 0: #ensure there are species left to collect
            if pa > 0: #smaller distances more likely
                probabilities = max_p - (sp_dists - min(sp_dists)) * (max_p - min_p) / (max(sp_dists) - min(sp_dists))
            elif pa == 0: #no phylogenetic structure
                probabilities = 0.5
            else: #greater distances more likely
                probabilities = min_p + (sp_dists - min(sp_dists)) * (max_p - min_p) / (max(sp_dists) - min(sp_dists))

            #selecting the next species based on probability
            next_sp = np.random.choice(dm.columns, p = probabilities)
            comm.append(next_sp)
            unused.remove(next_sp)
        
        else:
            raise Exception("No more species")
            break

    #Creating the presence/absence matrix
    if verbose:
        print(comm)

    
```

```python
#Testing the function
simcom(sr = 10, pa = 0.3, verbose = True)
```
